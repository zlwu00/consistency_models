```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.stats import gaussian_kde
import matplotlib.cm as cm

# Define the ODE
def ode_function(t, x, z):
    return (x - (x + t*z)) / t

# Time range for the integration
t_range = (0.1, 50)  # Adjusted time range
t_eval = np.linspace(t_range[0], t_range[1], 300)  # Time points where the solution is computed

# Initial condition
x0 = [0]

# Generate different Gaussian noise values and solve the ODE
for i in range(50):  # 50 different values of z
    z = np.random.normal()  # Gaussian noise
    sol = solve_ivp(ode_function, t_range, x0, t_eval=t_eval, args=(z,))
    plt.plot(sol.t, sol.y[0], color="gray")  # Trajectories in gray

# Adding labels and title
plt.xlabel('Time t')
plt.ylabel('x')
plt.title('Solutions of the ODE')

# Show the plot
plt.show()
```
